```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = [2, 0]
B = [1,1]

x_a, y_a = A[0], A[1]
x_b, y_b = B[0], B[1]
```

```python
m = (y_b - y_a) / (x_b - x_a) #coefficiente angolare della retta passante per a e b
#y = mx + q 
#y - y_a = m (x - x_a)
#q = y_a - m * x_a
q = y_b - m*x_b 
```

```python
def func (x, m, q):     
     return m * x + q 
    
def euclid (a, b) :
    delta_x = (b[0] - a[0])
    delta_y = (b[1]- a[1])
    distanza = np.sqrt (delta_x**2 + delta_y**2)
    return distanza 
```

```python
print ('coefficiente angolare:', m)
print ('intercetta:', q)
print ('equazione della retta: y =', m,'x + ', q)
```

```python
fig, ax = plt.subplots (nrows =1, ncols = 1)
x_grid = np.linspace (-3*m, 3*m, 1000)
y_grid = func (x_grid, m, q)

ax.plot(x_a, y_a, 'o', color='orange')
ax.plot(x_b, y_b, 'o', color='green')
ax.annotate (text = '  A', xy = (x_a, y_a))
ax.annotate (text = '  B', xy = (x_b, y_b))
ax.axhline (y = 0, color = 'grey')
ax.axvline (color = 'grey')
ax.plot (x_grid, y_grid)
plt.show()
```
